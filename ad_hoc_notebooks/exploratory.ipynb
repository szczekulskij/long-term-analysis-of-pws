```python
def new_func(metric = "total_GCE", time_break = 180):
    def save_metric_gen(series):
        try : 
            return float(series)
        except:
            return "None"

    df = get_data_df(metric = metric)
    output_df = pd.DataFrame(columns = ["surname","1st_visit", "2nd_visit", "last_visit_before_break", "break_time", "1st_visit_after_break","2nd_visit_after_break", "break_time_between_new_visits",])
    for surname in df.surname.unique():
        sub_df = df.loc[df.surname == surname]
        max_break = sub_df.time.max()
        if max_break >= time_break:
            break_visit_nr = int(sub_df.loc[sub_df.time == max_break].visit_nr)
            data = {}
            data["surname"] = surname
            data["1st_visit"] = save_metric_gen(sub_df.loc[sub_df.visit_nr == 1][metric])
            data["2nd_visit"] = save_metric_gen(sub_df.loc[sub_df.visit_nr == 1][metric])
            data["last_visit_before_break"] = save_metric_gen(sub_df.loc[sub_df.visit_nr == break_visit_nr-1][metric])
            data["break_time"] = max_break
            data["1st_visit_after_break"] = save_metric_gen(sub_df.loc[sub_df.visit_nr == break_visit_nr][metric])
            data["break_time_between_new_visits"] = save_metric_gen(sub_df.loc[sub_df.visit_nr == break_visit_nr+1].time)
            data["2nd_visit_after_break"] = save_metric_gen(sub_df.loc[sub_df.visit_nr == break_visit_nr+1][metric])
        output_df = output_df.append(data, ignore_index = True)
    output_df.drop_duplicates(inplace=True)
    return output_df
```
